```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def random_walk3d(steps):
    x, y, z = 0, 0, 0
    x_path, y_path, z_path = [x], [y], [z]

    for _ in range(steps):
        azimuth = np.random.uniform(-np.pi, np.pi)
        elevation = np.random.uniform(-np.pi/2, np.pi/2)
        distance = np.random.uniform(0, 1)
        x += distance * np.cos(elevation) * np.cos(azimuth)
        y += distance * np.cos(elevation) * np.sin(azimuth)
        z += distance * np.sin(elevation)
        x_path.append(x)
        y_path.append(y)
        z_path.append(z)

    return x_path, y_path, z_path
```

```python
def show_walk(x, y, z):
    fig = plt.figure(figsize=(10, 10))
    ax = fig.add_subplot(111, projection='3d')
    ax.plot(x, y, z, linestyle='-', marker='o', markersize=3)
    ax.set_title('3D Random Walk')
    ax.set_xlabel('X Coordinate')
    ax.set_ylabel('Y Coordinate')
    ax.set_zlabel('Z Coordinate')
    plt.show()
```

```python
steps = 250
```

```python
x_path, y_path, z_path = random_walk3d(steps)
show_walk(x_path, y_path, z_path)
```
