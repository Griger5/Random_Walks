```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def random_walk(steps):
    x = 0
    y = 0
    x_path = [0]
    y_path = [0]
    time = [0]

    for i in range(steps):
        angle = np.random.uniform(0, 2*np.pi)
        distance = np.random.uniform(0, 1)

        x += distance * np.cos(angle)
        y += distance * np.sin(angle)

        x_path.append(x)
        y_path.append(y)
        time.append(i)

    return x_path, y_path, time
```

```python
def show_walk(x, y, t):
    fig = plt.figure(figsize=(10, 10))
    show_walk = fig.add_subplot(111, projection="3d")
    show_walk.plot(x, y, t, linestyle="-", marker="o", markersize=3)
    show_walk.set_title("Random Walk on a Plane")
    show_walk.set_xlabel("X Coordinate")
    show_walk.set_ylabel("Y Coordinate")
    show_walk.set_zlabel("Time")
    show_walk.grid(True)
    plt.show()
```

```python
steps = 100
```

```python
x_path, y_path, time =  random_walk(steps)
show_walk(x_path, y_path, time)
```
